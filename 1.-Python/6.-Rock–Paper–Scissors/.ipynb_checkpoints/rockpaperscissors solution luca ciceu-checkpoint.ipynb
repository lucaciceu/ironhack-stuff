```python
from random import choice
import math

gestures = ['paper', 'rock', 'scissors']


def winner(player1_choice, player2_choice):

    choices = sorted([player1_choice, player2_choice])

    winner_choice = None
    if choices == ['paper', 'scissors']:
        winner_choice = 'scissors'
    if choices == ['paper', 'rock']:
        winner_choice = 'paper'
    if choices == ['rock', 'scissors']:
        winner_choice = 'rock'

    # tie
    if player1_choice == player2_choice:
        winner_player = 0

    if player1_choice == winner_choice:
        winner_player = 1

    if player2_choice == winner_choice:
        winner_player = -1

    return winner_player


# ask for how many rounds
n_rounds = 0
while not n_rounds % 2:
    try:
        n_rounds = int(input('How many rounds do you want to play (odd numbers)? '))
    except:
        pass

to_win = math.floor(n_rounds / 2) + 1

# if player wins game 1, if machine wins -1, 0 tie
rounds_score = {'human': 0, 'machine': 0, 'tie': 0}
rounds = 1
while not rounds > n_rounds:
    print(f'Round {rounds}')

    human_choice = None
    while human_choice not in gestures:
        human_choice = str.lower(input('\t{}? '.format('/'.join(gestures))))

    machine_choice = choice(gestures)
    print(f'\tHuman choice is {human_choice}, Machines choice is {machine_choice}.')

    # human is 1, machine is -1, tie is 0
    if winner(human_choice, machine_choice) == 0:
        rounds_score['tie'] = rounds_score['tie'] + 1
        print('\tTie, no winner for this round!')

    if winner(human_choice, machine_choice) == 1:
        rounds_score['human'] = rounds_score['human'] + 1
        print('\tHuman has won this round!')

    if winner(human_choice, machine_choice) == -1:
        rounds_score['machine'] = rounds_score['machine'] + 1
        print('\tMachine has won this round!')

    if rounds_score['human'] == to_win or rounds_score['machine'] == to_win:
        print('{} is game winner!!!'.format('HUMAN' if rounds_score['human'] == to_win else 'MACHINE'))
        break
    
    if rounds_score['tie'] == to_win:
        break
    
    rounds = rounds + 1

if rounds_score['tie'] == to_win or (rounds_score['human'] == rounds_score['machine'] and rounds_score['human'] < to_win):
    print('Game was a tie!')
```
