```python
import math

# robin
def euclidian_distance(point1, point2):

    point1_x = point1[0]
    point2_x = point2[0]
    point1_y = point1[1]
    point2_y = point2[1]

    return math.sqrt(pow(point2_x - point1_x, 2) + pow(point2_y - point1_y, 2))


# point = (x, y)
points = [
    (4, 5),
    (-0, 2),
    (4, 7),
    (1, -3),
    (3, -2),
    (4, 5),
    (3, 2),
    (5, 7),
    (-5, 7),
    (2, 2),
    (-4, 5),
    (0, -2),
    (-4, 7),
    (-1, 3),
    (-3, 2),
    (-4, -5),
    (-3, 2),
    (5, 7),
    (5, 7),
    (2, 2),
    (9, 9),
    (-8, -9),
]

# find arrow on arrow
arrow_on_arrow = []
for arrow_idx, arrow_point in enumerate(points):

    # only look for arrow on arrow in previous arrow points.
    if arrow_point in points[0:arrow_idx]:
        arrow_on_arrow.append(arrow_point)

print(f'There are {arrow_on_arrow} arrow(s) on arrow(s)')

# arrows in quadrants
quadrants = {'Q1': 0, 'Q2': 0, 'Q3': 0, 'Q4': 0}
for arrow_x, arrow_y in points:

    if arrow_x == 0 and arrow_y == 0:
        print(f'Arrow ({arrow_x}, {arrow_y}) is on center')
        continue

    # Q1
    if arrow_x > 0 and arrow_y > 0:
        quadrant = 'Q1'
    # Q2
    if arrow_x < 0 and arrow_y > 0:
        quadrant = 'Q2'
    # Q3
    if arrow_x < 0 and arrow_y < 0:
        quadrant = 'Q3'
    # Q4
    if arrow_x > 0 and arrow_y < 0:
        quadrant = 'Q4'

    quadrants[quadrant] = quadrants[quadrant] + 1
    print(f'Arrow ({arrow_x}, {arrow_y}) is on {quadrant} quadrant')

print('There are {}.'.format(', '.join([f'{arrow_count} arrow(s) in {quadrant}' for quadrant, arrow_count in quadrants.items()])))

# distance from center
distances = []
for point in points:
    distance = euclidian_distance(point, (0, 0))
    distances.append(distance)
    print(f'Arrow {point} is {distance:.2f} from center.')

min_distance = min(distances)
min_point_index = distances.index(min_distance)
print(f'Point {points[min_point_index]} as the minimum {min_distance} distance to center.')

# won't fit!
wont_fit = 0
for point in points:
    if euclidian_distance(point, (0, 0)) > 3:
        wont_fit = wont_fit + 1
print(f'There are arrows {wont_fit} which didn\'t fit!')      
```
