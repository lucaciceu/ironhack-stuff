```python
gandalf = [10, 11, 13, 30, 22, 11, 10, 33, 22, 22]
saruman = [23, 66, 12, 43, 12, 10, 44, 23, 12, 17]

gandalf_wins = 0
saruman_wins = 0
ties = 0

for gandalf_power, saruman_power in zip(gandalf, saruman):
    if gandalf_power > saruman_power:
        gandalf_wins = gandalf_wins + 1
    if saruman_power > gandalf_power:
        saruman_wins = saruman_wins + 1
    if saruman_power == gandalf_power:
        ties = ties + 1
        
print(f'Gandalf wins {gandalf_wins}')
print(f'Saruman wins {saruman_wins}')
print(f'Ties {ties}')
if gandalf_wins > saruman_wins:
    print('Gandalf has won!!!')
else:
    print('Saruman has won!!!')

#bonus
POWER = {'Fireball': 50, 'Lightning bolt': 40, 'Magic arrow': 10, 'Black Tentacles': 25, 'Contagion': 45}

gandalf_spells = [
    'Fireball',
    'Lightning bolt',
    'Lightning bolt',
    'Magic arrow',
    'Fireball',
    'Magic arrow',
    'Lightning bolt',
    'Fireball',
    'Black Tentacles',
    'Fireball',
    'Fireball',
]

saruman_spells = [
    'Contagion',
    'Contagion',
    'Black Tentacles',
    'Fireball',
    'Black Tentacles',
    'Lightning bolt',
    'Magic arrow',
    'Contagion',
    'Black Tentacles',
    'Magic arrow',
    'Magic arrow',
]

# compute matches
matches = []
for gandalf_spell, saruman_spell in zip(gandalf_spells, saruman_spells):

    gandalf_power = POWER[gandalf_spell]
    saruman_power = POWER[saruman_spell]

    # if Gandalf wins fight match is 1, if Saruman match is -1, match is 0 for tie
    match = 0
    if gandalf_power > saruman_power:
        match = 1
    if saruman_power > gandalf_power:
        match = -1

    matches.append(match)

# find winner
match_idx = 0
while True:
    # winner if 3 consecutives wins.
    winner_score = sum(matches[match_idx : match_idx + 3])

    # see if tie
    if winner_score in [-2, 2]:
        winner_score = sum(matches[match_idx : match_idx + 4])

    if winner_score in [-3, 3]:
        print('Winner of the battle is {}!'.format('Gandalf' if winner_score == 3 else 'Saruman'))
        break

    match_idx = match_idx + 3
    if match_idx > len(matches):
        print('No winner in this battle...')
        break
```
